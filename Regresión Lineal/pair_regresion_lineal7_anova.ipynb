```python
# Tratamiento de datos
# -----------------------------------------------------------------------
import numpy as np
import pandas as pd
import random 

# Gráficos
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Estadísticos
# ------------------------------------------------------------------------------
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.multivariate.manova import MANOVA
from sklearn.preprocessing import StandardScaler

plt.rcParams["figure.figsize"] = (10,8) 
```

```python
df = pd.read_csv('../datos/pokePd_estandarizado.csv', index_col = 0)
df.head()
```

```python
nuevas = {col: col.replace(".", "_")for col in df.columns}
df.rename(columns = nuevas, inplace = True)
```

```python
df.sample(5)
```

```python
lm = ols('Speed_BOX ~ Type + Total + HP + Attack + Defense + Sp_Atk + Sp_Def' , data=df).fit()
sm.stats.anova_lm(lm)
```

```python
lm.summary()
```

Interpretación ANOVA

1. Observamos que el R-Cuadrado (que es el porcentaje de la variable respuesta que explica las variables predictoras) es muy cercano a 1, por lo tanto las variables elegidas son correctas. Nuestras variables predictoras explican un 99% de la variación de la velocidad.

2. Pasamos a observar los p-valores de cada una de las variables (ignoraremos aquellas variables cuyos valores sean > 0.05) excepto en los tipos Hada, Eléctrico, Fantasma, Planta, psíquico, Roca y Acero, en el resto obtenemos valores menores de 0.05, lo que quiere decir que el resto de variables predictoras van a tener un efecto significativo sobre nuestra variable respuesta.

3. Respecto al coeficiente observamos que en la mayoría de los casos la relación es positiva para todas las subvariables de tipo y el total de puntos, mientas que en las últimas (HP, Sp.Def, etc...) se observa que la relación es inversa, es decir, cuando aumenta el valor de la variable predictora, disminuye el de la variable respuesta (y viceversa).
