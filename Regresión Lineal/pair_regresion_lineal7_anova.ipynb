```python
# Tratamiento de datos
# -----------------------------------------------------------------------
import numpy as np
import pandas as pd
import random 

# Gr谩ficos
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Estad铆sticos
# ------------------------------------------------------------------------------
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.multivariate.manova import MANOVA
from sklearn.preprocessing import StandardScaler

plt.rcParams["figure.figsize"] = (10,8) 
```

```python
df = pd.read_csv('../datos/pokePd_estandarizado.csv', index_col = 0)
df.head()
```

```python
nuevas = {col: col.replace(".", "_")for col in df.columns}
df.rename(columns = nuevas, inplace = True)
```

```python
df.sample(5)
```

```python
lm = ols('Speed_BOX ~ Type + Total + HP + Attack + Defense + Sp_Atk + Sp_Def' , data=df).fit()
sm.stats.anova_lm(lm)
```

```python
lm.summary()
```

Interpretaci贸n ANOVA
